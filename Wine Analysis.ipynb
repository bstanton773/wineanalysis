## Which types of wines&mdash;red or white&mdash;are associated with higher quality? 

```python
wines = ['red', 'gray']
wine_averages = df_wine.groupby('color').quality.mean()
wine_averages.plot(kind='bar', color=wines)
plt.title('Average Wine Quality by Color', fontsize=16)
plt.xlabel('Colors', fontsize=16)
plt.ylabel('Quality', fontsize=16);
```

```python
bin_edges = [8.00, 10.40, 14.90]
bin_names = ['low', 'high']
df_wine['booze_level'] = pd.cut(df_wine['alcohol'], bin_edges, labels=bin_names)
df_wine.tail()
```

```python
booze_averages = df_wine.groupby('booze_level')['quality'].mean()
booze_averages.plot(kind='barh', color=['green', 'pink'], figsize=(12,15))
plt.title('Average Wine Quality by Booze Level (10.4)', fontsize=16)
plt.xlabel('Booze Level', fontsize=16)
plt.ylabel('Quality', fontsize=16);
```
